```python
for k, v in b.items():
    ix1 = stoi_1[k[0]]
    ix2 = stoi_2[k[1]]
    N[ix1, ix2] = v
```

```python
# calcuate the P matrix
P = N.float()
P = P / P.sum(1, keepdim=True)
```

```python
P[0].sum()
```

```python
# sampling from the model
g = torch.Generator().manual_seed(2147483647)

for i in range(20):
    out = []
    ix1 = 0

    while True:
        p = P[ix1]
        # p = N[ix1].float()
        # p = p / p.sum()

        ix2 = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos_2[ix2])

        if ix2 == 0:
            break

        # update ix1
        cur_str = itos_1[ix1]
        print(cur_str)
        next_str = ""
        if len(cur_str) == 2:
            next_str = cur_str[1] + itos_2[ix2]
        # if cur_str == '.'
        else:
            next_str = cur_str + itos_2[ix2]

        ix1 = stoi_1[next_str]

    print(''.join(out))
```

```python
# evaluate this model

log_likelihood = 0.0
n = 0

for w in words:
    chs = ['.'] + list(w) + ['.']
    chs_temp = chs[2:]
    for i in range(len(chs_temp)):
        ch1 = chs[i] + chs[i+1]
        ch2 = chs_temp[i]
        ix1 = stoi_1[ch1]
        ix2 = stoi_2[ch2]
        prob = P[ix1, ix2]
        logprob = torch.log(prob)
        log_likelihood += logprob
        n += 1
        # print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')

print(n)
print(f'{log_likelihood=}')
nll = -log_likelihood
print(f'{nll=}')
print(f'{nll/n}')
```

```python
nlls = torch.zeros(4)
for i in range(4):
  # i-th bigram:
    x = xs[i].item() # input character index
    y = ys[i].item() # label character index
    print('--------')
    print(f'bigram example {i+1}: {itos_1[x]}{itos_2[y]} (indexes {x},{y})')
    print('input to the neural net:', x)
    print('output probabilities from the neural net:', probs[i])
    print('label (actual next character):', y)
    p = probs[i, y]
    print('probability assigned by the net to the the correct character:', p.item())
    logp = torch.log(p)
    print('log likelihood:', logp.item())
    nll = -logp
    print('negative log likelihood:', nll.item())
    nlls[i] = nll

print('=========')
print('average negative log likelihood, i.e. loss =', nlls.mean().item())
```

```python
# create training set
xs, ys = [], []

for w in words:
    chs = ['.'] + list(w) + ['.']
    chs_temp = chs[2:]
    for i in range(len(chs_temp)):
        ch1 = chs[i] + chs[i+1]
        ch2 = chs_temp[i]
        ix1 = stoi_1[ch1]
        ix2 = stoi_2[ch2]
        xs.append(ix1)
        ys.append(ix2)
        
xs = torch.tensor(xs)
ys = torch.tensor(ys)
num = xs.nelement()
print('number of examples: ', num)

# initialize the 'network'
g = torch.Generator().manual_seed(2147483647)
W = torch.randn((602, 27), generator=g, requires_grad=True)
```

```python
# gradient descent
for k in range(100):
  
    # forward pass
    xenc = F.one_hot(xs, num_classes=602).float() # input to the network: one-hot encoding
    logits = xenc @ W # predict log-counts
    counts = logits.exp() # counts, equivalent to N
    probs = counts / counts.sum(1, keepdims=True) # probabilities for next character
    loss = -probs[torch.arange(num), ys].log().mean()
    print(loss.item())
  
    # backward pass
    W.grad = None # set to zero the gradient
    loss.backward()
  
    # update
    W.data += -100 * W.grad
```
